```python
import sys
sys.path.append('..') #to add top-level to path

from modules.project_helper import VolFeatures
import matplotlib.pyplot as plt
```

```python
vol = VolFeatures()
vol.features('ES').head()
```

```python
fig, ax = plt.subplots(5,3,figsize=(15,20))
ax = ax.ravel()
for i, inst in enumerate(vol.instrument_list):
    ax[i].plot(vol.features(inst)[inst + '_1M_atm_vol'])
    ax[i].set_title('1M atm volatility, {}'.format(inst))
```

```python
fig, ax = plt.subplots(5,3,figsize=(15,20))
ax = ax.ravel()
for i, inst in enumerate(vol.instrument_list):
    ax[i].plot(vol.features(inst)[inst + '_2M_1M_atm_vol'])
    ax[i].set_title('2M - 1M atm volatility, {}'.format(inst))
```

```python
fig, ax = plt.subplots(5,3,figsize=(15,20))
ax = ax.ravel()
for i, inst in enumerate(vol.instrument_list):
    ax[i].plot(vol.features(inst)[inst + '_1M_RR25'], label = '1M')
    ax[i].plot(vol.features(inst)[inst + '_2M_RR25'], label = '2M')
    ax[i].set_title('25 Delta Risk Reversal, {}'.format(inst))
    ax[i].legend();
```
